### Decision Tree Classification

### Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Importing the datasets

```python
import sklearn.datasets as datasets
iris=datasets.load_iris()
data=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
data.head()
```

```python
X=data
y=iris.target
```

```python
X.head()
```

```python
X.describe()
```

```python
classes=iris.target_names
classes
```

```python
y
```

```python
sns.pairplot(X)
```

### Splitting the data into the Train and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train.head()
```

```python
y_train
```

### Training the Decision Tree Classifier Model 

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',random_state=0)
model.fit(X_train,y_train)
```

### Making Predictions

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_pred.reshape(len(y_test),1)),1))
```

### Evaluating the model

```python
from sklearn import metrics  
print('Acuuracy score:', 
      metrics.accuracy_score(y_test, y_pred) )
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred) )
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```

### Visualising the Decision Tree

```python
from sklearn import tree
plt.figure(figsize=(30,25))
visual=tree.plot_tree(model, feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```
